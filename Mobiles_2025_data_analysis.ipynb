```python
df['Mobile Weight'] = df['Mobile Weight'].astype(str).str.replace('g', '', regex=False).astype(float)
df['Launched Price (Pakistan)'] = df['Launched Price (Pakistan)'].astype(str).str.replace('PKR', '', regex=False).str.replace(',', '', regex=False).str.strip().replace('Not available', np.nan).astype(float)
df['Launched Price (India)'] = df['Launched Price (India)'].astype(str).str.replace('INR', '', regex=False).str.replace(',', '', regex=False).str.strip().replace('Not available', np.nan).astype(float)
df['Launched Price (China)'] = df['Launched Price (China)'].astype(str).str.replace('CNY', '', regex=False).str.replace('¥', '', regex=False).str.replace(',', '', regex=False).str.strip().replace('Not available', np.nan).astype(float)
df['Launched Price (USA)'] = df['Launched Price (USA)'].astype(str).str.replace('USD', '', regex=False).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip().replace('Not available', np.nan).astype(float)
df['Launched Price (Dubai)'] = df['Launched Price (Dubai)'].astype(str).str.replace('AED', '', regex=False).str.replace(',', '', regex=False).str.strip().replace('Not available', np.nan).astype(float)
```
